#### a) Obliczanie wartości wybranych opisowych charakterystyk rozkładu

Korzystając z danych zawartych w tablicy `recession_table_1` oblicz dla każdego roku następujące opisowe charakterystyki rozkładu:
- wartość średnia (ang. mean),
- odchylenie standardowe (ang. standart deviation),
- kwartyle (ang. quartiles),
- moda (ang. mode).

Zestaw wyniki w tablicy `descriptive_parameters_2`.

Indeks:
 - zawiera następujące nazwy wierszy: mean, std, Q1, Q2, Q3, mode.
 
Kolumny:
 - lata (osobna kolumna dla każdego roku opatrzona nagłówkiem 2006, 2007 itd. sformatowanym jako str)
 
Uwaga! W celu otrzymania maksymalnej liczby punktów otrzymane wyniki powinny być w całości wynikiem pracy napisanego kodu, bez manualnego określania, w którym przedziale znajdują się określone statystyki itp.

```python
# YOUR CODE HERE
years = ['2006','2007','2008','2009','2010','2011']
parameters = ['mean', 'std', 'Q1', 'Q2', 'Q3', 'mode']

descriptive_parameters_2 = pd.DataFrame(columns=years, index=parameters)

for year in years:
    year_data = recession_table_1[year]
    midpoints = [(interval.left + interval.right) / 2 for interval in year_data.index]
    # CALCULATE MEAN
    mean = np.sum(midpoints * year_data) / year_data.sum()
    # CALCULATE STANDARD DEVIATION
    std = np.sqrt(np.sum((midpoints - mean)**2 * year_data) / year_data.sum())
    # CALCULATE MODE
    nD = year_data.max()
    nD_index = year_data[year_data == nD].index
    xD = nD_index.tolist()[0].left
    nD_minus1 = year_data.shift(1).loc[nD_index].values[0]
    nD_plus1 = year_data.shift(-1).loc[nD_index].values[0]
    interval_length = nD_index.tolist()[0].length
    mode = xD + ((nD - nD_minus1) / ((nD - nD_plus1) + (nD - nD_minus1))) * interval_length
    # CALCULATE QUARTILE
    NQ1 = year_data.sum() / 4
    NQ2 = year_data.sum() / 2
    NQ3 = 3 * year_data.sum() / 4 
    # Q1
    year_data_cumsum = year_data.cumsum()
    q1_index = year_data[year_data_cumsum >= NQ1].index[0]
    q1_index_before = year_data[year_data_cumsum < NQ1].index[-1]
    xQ1 = q1_index.left
    nQ1 = year_data[q1_index]
    before = year_data_cumsum[q1_index_before]
    Q1 = xQ1 + (NQ1 - before) * interval_length / nQ1
    # Q2
    q2_index = year_data[year_data_cumsum >= NQ2].index[0]
    q2_index_before = year_data[year_data_cumsum < NQ2].index[-1]
    xQ2 = q2_index.left
    nQ2 = year_data[q2_index]
    before = year_data_cumsum[q2_index_before]
    Q2 = xQ2 + (NQ2 - before) * interval_length / nQ2
    # Q3
    q3_index = year_data[year_data_cumsum >= NQ3].index[0]
    q3_index_before = year_data[year_data_cumsum < NQ3].index[-1]
    xQ3 = q3_index.left
    nQ3 = year_data[q3_index]
    before = year_data_cumsum[q3_index_before]
    Q3 = xQ3 + (NQ3 - before) * interval_length / nQ3
    # ADD TO DATA FRAME
    descriptive_parameters_2.loc['mean', year] = mean
    descriptive_parameters_2.loc['std', year] = std
    descriptive_parameters_2.loc['mode', year] = mode
    descriptive_parameters_2.loc['Q1', year] = Q1
    descriptive_parameters_2.loc['Q2', year] = Q2
    descriptive_parameters_2.loc['Q3', year] = Q3
    
descriptive_parameters_2
```

```python
### Komórka testowa
# Testy wybranych wartości z tablicy
assert np.isclose(descriptive_parameters_2.loc["mean", "2006"], 5.606061)
assert np.isclose(descriptive_parameters_2.loc["std", "2007"], 4.184001)
assert np.isclose(descriptive_parameters_2.loc["Q1", "2008"], 0.916667)
assert np.isclose(descriptive_parameters_2.loc["Q2", "2009"], 0.220588)
assert np.isclose(descriptive_parameters_2.loc["Q3", "2010"], 7.604167)
assert np.isclose(descriptive_parameters_2.loc["mode", "2011"], 3.691589)
```
