#### b) Charakterystyka rozkładów
Na podstawie przygotowanych wykresów scharakteryzuje opisane rozkłady teoretyczne i odpowiedz na pytanie w jaki sposób zmiana wartości parametrów rozkładów wpływa na położenie i kształt ich funkcji gęstości prawdopodobieństwa.

Zmiana wartości parametrów rozkładów wpływa na położenie i kształt funkcji gęstości prawdopodobieństwa różnych rozkładów prawdopodobieństwa w różny sposób. Przedstawmy to bardziej szczegółowo dla poszczególnych rozkładów:

- Rozkład normalny: Zmiana wartości średniej i wariancji wpływa na położenie i kształt funkcji gęstości prawdopodobieństwa rozkładu normalnego. Zwiększenie wartości średniej przesuwa funkcję gęstości prawdopodobieństwa w prawo, natomiast zmniejszenie wartości średniej przesuwa funkcję w lewo. Zwiększenie wartości wariancji powoduje rozproszenie wartości wokół średniej, co prowadzi do spłaszczenia funkcji gęstości, natomiast zmniejszenie wartości wariancji powoduje zwiększenie skupienia wartości wokół średniej, co prowadzi do bardziej wąskiej i wyższej funkcji gęstości prawdopodobieństwa.

- Rozkład t-Studenta: Zmiana stopni swobody wpływa na położenie i kształt funkcji gęstości prawdopodobieństwa rozkładu t-Studenta. Im większa liczba stopni swobody, tym bardziej funkcja gęstości prawdopodobieństwa przypomina funkcję gęstości rozkładu normalnego. Im mniejsza liczba stopni swobody, tym bardziej spłaszczona jest funkcja gęstości, co oznacza większą zmienność danych.

- Rozkład chi-kwadrat (chi^2): Zmiana liczby stopni swobody wpływa na położenie i kształt funkcji gęstości prawdopodobieństwa rozkładu chi-kwadrat. Im większa liczba stopni swobody, tym bardziej funkcja gęstości przypomina rozkład normalny. Im mniejsza liczba stopni swobody, tym bardziej spłaszczona jest funkcja gęstości.

- Rozkład F-Snedecor: Zmiana liczby stopni swobody w liczniku i mianowniku wpływa na położenie i kształt funkcji gęstości prawdopodobieństwa rozkładu F-Snedecor. Im większa liczba stopni swobody, tym bardziej funkcja gęstości przypomina rozkład normalny. Im mniejsza liczba stopni swobody, tym bardziej spłaszczona jest funkcja gęstości. Przy zwiększaniu liczby stopni swobody w liczniku i mianowniku jednocześnie, funkcja gęstości staje się węższa i wyższa.

#### a) Wyliczanie prawdopodobieństw na podstawie wartości

Oblicz i zapisz do zmiennych o podanych nazwach następujące wartości prawdopodobieństwa:
- `p1` - $P(X < -1.5)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład t-Studenta o 10 stopniach swobody - $t(10)$,
- `p2` - $P(X > 2)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład F Snedecora o 5 i 50 stopniach swobody - $F(5, 50)$,
- `p3` - $P(3 < X < 10)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład chi-kwadrat o 4 stopniach swobody - $\chi^2(4)$,
- `p4` - $P(X < -2.75 | X > 1.5)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład normalny o wartości oczewiwanej równej -0.5 i odchyleniu standardowym równym 1.5 - $N(-0.5, 1.5)$,

Każde z rozwiązań zilustruj wykresem zawierającym krzywą gęstości prawdopodobieństwa rozkładu z zaznaczonym kolorem fragmentem pola pod krzywą odpowiadający wyznaczonej wartości prawdopodobieństwa. Przygotuj figurę zgodnie z wytycznymi z nagłówka notatnika. Dodatkowo:
- ustaw zakres osi x od wartości zmiennej $X$, dla której dystrybuanta analizowanego rozkładu przyjmuje wartość 0.0001 do wartości zmiennej $X$, dla której dystrybuanta analizowanego rozkładu przyjmuje wartość 0.9999,
- w tytule wykresu zawrzyj uzupełniony wzór na prawdopodobieństwo uzupełniony o wyliczoną wartość prawdopodobieństwa podaną do 4 miejsca po przecinku (np. P(X < -1.5) = x.xxxx, gdzie x.xxxx to obliczona wartość).

```python
# p1
# YOUR CODE HERE
t_dist = sp.stats.t(10)
p1 = t_dist.cdf(-1.5)
x = np.linspace(t_dist.ppf(0.0001), t_dist.ppf(0.9999), 1000)
fig, ax = plt.subplots(1,1,figsize=(6,6))
fig.set_facecolor('white')
ax.plot(x, t_dist.pdf(x), color='red')
ax.fill_between(x[x<-1.5], t_dist.pdf(x[x<-1.5]), 0, alpha=0.4)
ax.set_title(f'P(X < -1.5) = {p1:.4f}')
ax.vlines(x=-1.5, ymin=0, ymax=t_dist.pdf(-1.5), linestyle='--', color='lightblue')
ax.set_xlabel('x')
ax.set_ylabel('p(x)')
```

```python
# p2
# YOUR CODE HERE
f_dist = sp.stats.f(5, 50)
p2 = 1 - f_dist.cdf(2)
x = np.linspace(f_dist.ppf(0.0001), f_dist.ppf(0.9999), 1000)
fig, ax = plt.subplots(1,1,figsize=(6,6))
fig.set_facecolor('white')
ax.plot(x, f_dist.pdf(x), color='red')
ax.fill_between(x[x>2], f_dist.pdf(x[x>2]), 0, alpha=0.4, color='lightblue')
ax.set_title(f'P(X > 2) = {p2:.4f}')
ax.vlines(x=2, ymin=0, ymax=f_dist.pdf(2), linestyle='--')
ax.set_xlabel('x')
ax.set_ylabel('p(x)')
```

```python
# p3
# YOUR CODE HERE
chi2_dist = sp.stats.chi2(4)
p3 = chi2_dist.cdf(10) - chi2_dist.cdf(3)
x = np.linspace(chi2_dist.ppf(0.0001), chi2_dist.ppf(0.9999), 1000)
fig, ax = plt.subplots(1,1,figsize=(6,6))
fig.set_facecolor('white')
ax.plot(x, chi2_dist.pdf(x), color='red')
ax.fill_between(x[(x >= 3) & (x <= 10)], chi2_dist.pdf(x[(x >= 3) & (x <= 10)]), 0, alpha=0.4, color='lightblue')
ax.set_title(f'P(3 < X < 10) = {p3:.4f}')
ax.vlines(x=3, ymin=0, ymax=chi2_dist.pdf(3), linestyle='--')
ax.vlines(x=10, ymin=0, ymax=chi2_dist.pdf(10), linestyle='--')
ax.set_xlabel('x')
ax.set_ylabel('p(x)')
```

```python
# p4
# YOUR CODE HERE
norm_dist = sp.stats.norm(-0.5, 1.5)
p4 = norm_dist.cdf(-2.75) + (1 - norm_dist.cdf(1.5))
x = np.linspace(norm_dist.ppf(0.0001), norm_dist.ppf(0.9999), 1000)
fig, ax = plt.subplots(1,1,figsize=(6,6))
fig.set_facecolor('white')
ax.plot(x, norm_dist.pdf(x), color='red')
ax.fill_between(x[(x <= -2.75)], norm_dist.pdf(x[(x <= -2.75)]), 0, alpha=0.4, color='lightblue')
ax.fill_between(x[(x >= 1.5)], norm_dist.pdf(x[(x >= 1.5)]), 0, alpha=0.4, color='lightblue')
ax.set_title(f'P(X < -2.75 | X > 1.5) = {p4:.4f}')
ax.vlines(x=-2.75, ymin=0, ymax=norm_dist.pdf(-2.75), linestyle='--')
ax.vlines(x=1.5, ymin=0, ymax=norm_dist.pdf(1.5), linestyle='--')
ax.set_xlabel('x')
ax.set_ylabel('p(x)')
```

#### b) Wyliczanie wartości na podstawie prawdopodobieństw

- `s3` - $P(-1.25 < X < s) = 0.35$, gdzie x to zmienna losowa ciągła posiadająca rozkład normalny o wartości oczewiwanej równej -0.4 i odchyleniu standardowym równym 0.75 - $N(-0.4, 0.75)$,

Każde z rozwiązań zilustruj wykresem zawierającym krzywą gęstości prawdopodobieństwa rozkładu z zaznaczonym kolorem fragmentem pola pod krzywą odpowiadający wartości prawdopodobieństwa określonej przez rozwiązywane równanie. Przygotuj figurę zgodnie z wytycznymi z nagłówka notatnika. Dodatkowo:
